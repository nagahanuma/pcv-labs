3.Implement a code to perform blending operation on two images in which have 60% of image1 and 40% of image2.

```python
#Operation to perform in blending position of image1 and image2 is 60%,40%
import cv2
import numpy as np
image1 = cv2.imread("hills.jpg")
image2 = cv2.imread("dog.jpg")
print("image1 size : ", image1.shape)
print("image2 size : ", image2.shape)

scale_percent=60

width = int(image1.shape[1] * scale_percent/100)
height = int(image1.shape[1] * scale_percent/100)

width = int(image2.shape[1] * scale_percent/100)
height = int(image2.shape[1] * scale_percent/100)

dim = (width, height)

resized1 = cv2.resize(image1, dim, interpolation=cv2.INTER_AREA)
resized2 = cv2.resize(image2, dim, interpolation=cv2.INTER_AREA)
                      
print("image1 resized :", resized1.shape)
print("image2 resized :", resized2.shape)

cv2.imshow("Resize image",resized1)
cv2.imshow("Resize image",resized2)
cv2.waitKey()
cv2.destroyAllWindows()


difference = cv2.subtract(resized1, resized2)
                      
result = not np.any(difference)
                      
if result is True:
    print("The images are the same")
else:
    cv2.imwrite("result.jpg",difference)
    print("The images are different")
```
