# Edge Detection and Region Of Interest Using OpenCV and Python

# IN-LAB TASK:-

```python
#1
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("marvel.jpg")
plt.figure(figsize =(10, 10))
plt.subplot(2,2,1),plt.imshow(img)
plt.title('Original'), plt.xticks([]), plt.yticks([])
```

```python
# converting to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize =(10, 10))
plt.subplot(2,2,2),plt.imshow(gray,cmap = 'gray')
plt.title('Grey Scale'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# remove noise
img = cv2.GaussianBlur(img,(45,45),0)
plt.figure(figsize =(10, 10))
plt.subplot(2,2,3),plt.imshow(img)
plt.title('Gaussian Blur'), plt.xticks([]), plt.yticks([])
```

```python
img = cv2.imread("marvel.jpg")
# converting to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)

laplacian = cv2.Laplacian(img,cv2.CV_64F)
plt.figure(figsize =(10, 10))
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.show()
```
