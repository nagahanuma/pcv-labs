```python
from tkinter import *
import cv2
from PIL import ImageTk, Image
from tkinter import filedialog
import matplotlib.pyplot as plt 
```

```python
def openfilename():
    filename = filedialog.askopenfilename(title ='"pen')
    return filename
def open_img():
    x = openfilename()
    img = Image.open(x)
    img = img.resize((250, 250), Image.ANTIALIAS)
    img = ImageTk.PhotoImage(img)
    i = Label(root,image=img)
    i.image=img
    i.pack()
    canvas1.create_image(20,30,anchor='nw',image=i)
    print("hello")
def apply_binary():
    print("binary")
def apply_binary_inv():
    print("binary_inv")
def apply_tozero():
    print("tozero")
def apply_tozero_inv():
    print("Tozero Inv")
def apply_trunc():
    print("Trunc")
def apply_mean_c():
    print("mean_c")
def apply_gaussian_c():
    print("gaussian_c")
```

```python
root = Tk()
canvas1 = Canvas(root, width = 800, height = 600)
canvas1.pack()
label1 = Label(root, text='Graphical User Interface')
label1.config(font=('Arial', 20))
canvas1.create_window(400, 50, window=label1)

button1 = Button (root, text=' Open Image ',command=open_img, bg='palegreen2', font=('Arial', 11, 'bold')) 
canvas1.create_window(400, 180, window=button1)

button2 = Button (root,text=' Quit ',command = root.destroy, bg='palegreen2', font=('Arial', 11, 'bold'))
canvas1.create_window(400,430, window=button2)

button3 = Button(root,text=" BINARY ",command = apply_binary, bg = 'lightskyblue2', font=('Arial', 11, 'bold'))
canvas1.create_window(400,220, window=button3)

button4 = Button(root,text=" BINARY_INV ",command = apply_binary_inv, bg = 'lightskyblue2', font=('Arial', 11, 'bold'))
canvas1.create_window(400,250, window=button4)

button5 = Button(root,text=" TOZERO ",command = apply_tozero, bg = 'lightskyblue2', font=('Arial', 11, 'bold'))
canvas1.create_window(400,280, window=button5)

button6 = Button(root,text=" TOZERO_INV ",command = apply_tozero_inv, bg = 'lightskyblue2', font=('Arial', 11, 'bold'))
canvas1.create_window(400,310, window=button6)

button7 = Button(root,text=" TRUNC ",command = apply_trunc, bg = 'lightskyblue2', font=('Arial', 11, 'bold'))
canvas1.create_window(400,340, window=button7)

button8 = Button(root,text=" MEAN_C ",command = apply_mean_c, bg = 'lightskyblue2', font=('Arial', 11, 'bold'))
canvas1.create_window(400,370, window=button8)

button9 = Button(root,text=" GAUSSIAN_C ",command = apply_gaussian_c, bg = 'lightskyblue2', font=('Arial', 11, 'bold'))
canvas1.create_window(400,400, window=button9)

root.mainloop()
```
